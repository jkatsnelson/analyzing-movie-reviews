```python
import pandas as pd

movies = pd.read_csv('fandango_score_comparison.csv')

movies
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.hist(movies['Fandango_Stars'])
```

```python
plt.hist(movies['Metacritic_norm_round'])
```

```python
fan = movies['Fandango_Stars']
meta = movies['Metacritic_norm_round']

fan_mean = fan.mean()
meta_mean = meta.mean()

fan_mode = fan.mode()
meta_mode = meta.mode()

import numpy

fan_std = numpy.std(fan)
meta_std = numpy.std(meta)

print(fan_mean, meta_mean, fan_mode, meta_mode, fan_std, meta_std)
```

```python
fm_diff = (fan - meta).abs()
fm_diff
movies['fm_diff'] = fm_diff
movies.sort_values(by="fm_diff", ascending=False).head(n=5)
```

```python
import scipy.stats as stats

stats.pearsonr(fan, meta)
```

```python
meta_fan_lr = stats.linregress(meta, fan)
def predict_fandango_score (s):
    return s * meta_fan_lr[0] + meta_fan_lr[1]

pred_3 = predict_fandango_score(3)
pred_3
```

```python
predict_fandango_score(1)
```

```python
predict_fandango_score(5)
```

```python
plt.xlabel('Metacritic')
plt.ylabel('Fandango')
# plt.scatter(meta, fan)
# plt.plot(3, predict_fandango_score(3))
plt.scatter(meta, fan)
plt.plot([1.0, 5.0], [predict_fandango_score(1.0), predict_fandango_score(5.0)])
plt.xlim([1, 5])
```

```python
meta, rot = movies['Metacritic_norm'], movies['RT_norm']
meta_rot_lr = stats.linregress(meta, rot)

def predict_rot_score (s):
    return s * meta_rot_lr[0] + meta_rot_lr[1]

plt.xlabel('Metacritic')
plt.ylabel('RottenTomatoes')

plt.scatter(meta, rot)
plt.plot([1.0, 5.0], [predict_rot_score(1.0), predict_rot_score(5.0)])
plt.xlim([1, 5])
```
